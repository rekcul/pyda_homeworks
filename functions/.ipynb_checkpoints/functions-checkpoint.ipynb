# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

#### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

````python
def find_person_by_number(number):
    """
    Пользователь по команде "p" может узнать владельца документа по его номеру
    
    Примеры работы:
    1.
    find_person_by_number()
    Введите номер документа:
    10006
    Результат:
    Владелец документа: Аристарх Павлов
    
    2.
    find_person_by_number()
    Введите номер документа:
    12345
    Результат:
    Документ не найден в базе
    """
    name = None
    for i in range(len(documents)):
        if documents[i]['number']==number:
            name=documents[i]['name']
    return name
    
````

#### Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

````python
def find_shelf_by_number(number):
    """
    Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    Пример работы:

    find_shelf_by_number()
    Введите номер документа:
    10006
    Результат:
    Документ хранится на полке: 2
    """
    shelf = None
    for i in list(directories.keys()):
        for j in range(len(directories[i])):
            if directories[i][j]==number:
                shelf=i
    return shelf
````

#### Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам

````python
def find_info_by_number():
    """
    Пример работы:
    Введите команду:
    l
    Результат:
    №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
    №: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
    №: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
    """
    nums = [i['number'] for i in documents]
    for i in nums:
        shelf = find_shelf_by_number(i)
        for j in range(len(documents)):
            if documents[j]['number']==i:
                print('№: {}, тип: {}, владелец: {}, полка хранения: {}'.format(i, documents[j]['type'], documents[j]['name'], shelf))
    return
````

#### Пункт 4. Пользователь по команде "ads" может добавить новую полку¶

````python
def add_new_folder(shelf_number):
    """
    Примеры работы:
    1.
    Введите команду:
    ads

    Введите номер полки:
    10
    Результат:
    Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.
    2.
    Введите команду:
    ads

    Введите номер полки:
    1
    Результат:
    Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

    """
    if shelf_number in list(directories.keys()):
        return 'nothing'
    else:
        directories[shelf_number]=[]        
        return 
````

#### Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)¶

````python
def delete_folder(shelf_number):
    """
    Примеры работы:
    1.
    Введите команду:
    ds

    Введите номер полки:
    3
    Результат:
    Полка удалена. Текущий перечень полок: 1, 2.
    2.
    Введите команду:
    ds

    Введите номер полки:
    1
    Результат:
    На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.
    3.
    Введите команду:
    ds

    Введите номер полки:
    4
    Результат:
    Такой полки не существует. Текущий перечень полок: 1, 2, 3.
    
    """
    
    if shelf_number in list(directories.keys()):
        if directories[shelf_number] != []:
            return 'cant delete'
        else:
            directories.pop(shelf_number)
            return 'deleted'
    else:
        return 'nothing'
````

````python
def secretary():
    """
    Пользователь по команде "p" может узнать владельца документа по его номеру
    Примеры работы:
    1.
    Введите команду:
    p

    Введите номер документа:
    10006
    Результат:
    Владелец документа: Аристарх Павлов
    2.
    Введите команду:
    p

    Введите номер документа:
    12345
    Результат:
    Документ не найден в базе
    """
    inp = input('Введите команду: ')
    
    while inp  != 'q':

        if inp  == 'p':
            number = input('Введите номер документа: ')
            name = find_person_by_number(number)
            if name is None:
                print('Документ не найден в базе')
            else:
                print('Владелец документа: {}'.format(name))
            inp = input('Введите команду: ')

        elif inp == 's':
            number = input('Введите номер документа: ')
            shelf = find_shelf_by_number(number)
            if shelf is None:
                print('Документ не найден в базе')
            else:
                print('Документ хранится на полке: {}'.format(shelf))
            inp = input('Введите команду: ')

        elif inp == 'l':
            ans = find_info_by_number()
            return ans

        elif inp == 'ads':
            shelf_number = input('Введите номер полки: ')
            ans = add_new_folder(shelf_number)
            if ans == 'nothing':
                print('Такая полка уже существует. Текущий перечень полок: {}'.format(','.join(list(directories.keys()))))
            else:
                print('Полка добавлена. Текущий перечень полок: {}'.format(','.join(list(directories.keys()))))
            inp = input('Введите команду: ')

        elif inp == 'ds':
            shelf_number = input('Введите номер полки: ')
            ans = delete_folder(shelf_number)
            if ans == 'nothing':
                print('Такой полки не существует. Текущий перечень полок: {}'.format(','.join(list(directories.keys()))))
            elif ans == 'deleted':
                print('Полка удалена. Текущий перечень полок: {}'.format(','.join(list(directories.keys()))))
            elif ans == 'cant delete':
                print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {}'.format(','.join(list(directories.keys()))))
            inp = input('Введите команду: ')
        
````

````python
secretary()
````
